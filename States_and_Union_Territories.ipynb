```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/States_and_union_territories_of_India'
page = requests.get(url)
soup = BeautifulSoup(page.content,'lxml')
#soup
```

```python
state_table = soup.find_all(class_="wikitable sortable plainrowheaders")[0]
union_table = soup.find_all(class_="wikitable sortable plainrowheaders")[1]
```

```python
heads = [h.get_text().strip() for h in state_table.find_all('th')]
heading = heads [:11]
states = heads [11:]
```

```python
tdata = state_table.find_all('tr')[1:]
tdata
```

```python
[d.get_text().strip() for d in tdata[0].find_all('td')]
```

```python
all_data = []
for dt in tdata:
    l = [d.get_text().strip() for d in dt.find_all('td')]
    all_data.append(l)
    
all_data
```

```python
iso,veh_cd,zone,cap,larcty,sthd,popl,area,offlang,add_offlang=[],[],[],[],[],[],[],[],[],[]

for q in all_data:
    if len(q)==9:
        iso.append(q[0])
        veh_cd.append(q[1])
        zone.append(q[2])
        cap.append(q[3])
        larcty.append(q[3])
        sthd.append(q[4])
        popl.append(q[5])
        area.append(q[6])
        offlang.append(q[7])
        add_offlang.append(q[8])
    else:
        iso.append(q[0])
        veh_cd.append(q[1])
        zone.append(q[2])
        cap.append(q[3])
        larcty.append(q[4])
        sthd.append(q[5])
        popl.append(q[6])
        area.append(q[7])
        offlang.append(q[8])
        add_offlang.append(q[9])
```

```python
p = [states,iso,veh_cd,zone,cap,larcty,sthd,popl,area,offlang,add_offlang]
d = dict(list(zip(heading,p)))
```
