## Union_Territories

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re

url = 'https://en.wikipedia.org/wiki/States_and_union_territories_of_India'
page = requests.get(url)
soup = BeautifulSoup(page.content,'lxml')

#state_table = soup.find_all(class_="wikitable sortable plainrowheaders")[0]
union_table = soup.find_all(class_="wikitable sortable plainrowheaders")[1]

#heads = [h.get_text().strip() for h in union_table.find_all('th')]
data = union_table.find_all('tr')

info=[]
heads = [h.get_text().strip() for h in data[0].find_all('th')]

for i in data[1:]:
    L=[d.get_text().strip() for d in i.find_all('td')]
    if re.match(r'^[0-9]+\s[A-z]*[a-z]*\s[0-9]+$', L[4]):
        L.insert(4,L[3])
    L.insert(0, *[d.get_text().strip() for d in i.find_all('th')])
    info.append(L)
    
#info = [h.get_text().strip() for h in data[1].find_all('td')]

```

```python
d = dict(list(zip(heads, zip(*info))))
df=pd.DataFrame(d)
df
```
